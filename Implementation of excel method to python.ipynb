```python
def weekpred2(data1,data2,give_delta=True):
    selectorprob=data1[(data1["ay"]==3)&(data1["hafta_durum"]=="hafta içi")]["hafta_4"].values[0]
    selectornum=data2[(data2["ay"]==3)&(data2["hafta_durum"]=="hafta içi")]["hafta_4"].values[0]
    selectorre=selectornum/selectorprob
    cols=data1.columns[2:7]
    valsprob=data1[cols].values
    newnums=valsprob*selectorre
    dff=data2.copy()
    dff[cols]=newnums
    return dff

```

```python
#data1=prop
#data2=number
#Asıl fonksiyon
#give_delta if set True gives a error dataframe
def weekpred(data1,data2,baseweek=1,give_delta=True):
    selector_col=data1.columns[baseweek+1]
    selector_series=data2[selector_col]/data1[selector_col]
    liste=[]
    listedelta=[]
    for i in data1.columns:
        if i ==selector_col or i=="ay" or data1[i].dtype=="O":
            liste.append(data2[i])
            listedelta.append(data2[i])
        else:
            predicted_series=selector_series*data1[i]
            delta_series=data2[i]-predicted_series
            liste.append(predicted_series)
            listedelta.append(delta_series)
    dataf=pd.DataFrame(liste,index=data1.columns).T
    datadelta=pd.DataFrame(listedelta,index=data1.columns).T
    if give_delta==True:
        return dataf,datadelta
    else:
        return dataf
    
#Supplementary   
def maed(datar,datap,start=2,stop=7):
    columns=datar.columns[start:stop]
    abserror=abs(datap[columns]-datar[columns]).sum().sum()
    total=datap[columns].size
    return abserror/total
    
def mape(datar,datap,start=2,stop=7):
    columns=datar.columns[start:stop]
    abserror=abs(datap[columns]-datar[columns])/datar[columns]
    percentageseries=abserror.sum().sum()
    total=datap[columns].size
    return percentageseries/total    
    
    
    
```
