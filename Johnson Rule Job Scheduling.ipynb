```python
def job_order_sequence(df,machine_order):
    gonnabe_df=[]
    sortii=[]
    sort_index=[]
    den=[]
    machine_copy=0
    summ=0
    for i in range(1,len(machine_order)):

        if df[machine_order[0]].min()>=df[machine_order[i]].max():
            machine_copy=machine_order.copy()
            machine_copy.remove(machine_order[i])
            summ=machine_order[i]
        elif df[machine_order[len(machine_order)-1]].min()>=df[machine_order[i]].max():
            machine_copy=machine_order.copy()
            machine_copy.remove(machine_order[i])
            summ=machine_order[i]

        elif machine_copy !=0:
            for a in machine_copy:
                den=df[summ] +df[a]
                den[den.size +1]=den.sum()
                den.name= a + "+" + summ
                gonnabe_df.append(den)
            machine_copy=0

            sortii=pd.DataFrame(gonnabe_df)
            sortii=sortii[sortii[len(den)]==sortii[len(den)].min()].T
            sort_index=sortii.sort_values(by=sortii.columns[0], ascending=True).index.tolist()
            sort_index.remove(len(den))
            return sort_index
        #Our goal was finding this index values of  sorting of jobs 
        elif i==len(machine_order)-1 and den==[]:
            print("Johnson Not Applicable!!!")
            continue        

    
```
